```python
chipo['item_price'] = chipo['item_price'].str.replace('$', '').astype(float)
```

```python
def categorize_price(Price):
    if Price < 5:
        return 'Low'
    elif Price < 10:
        return 'Medium'
    else:
        return 'High'
    
chipo['price_group'] = chipo['Price'].apply(categorize_price)
chipo.head()
```
